```python
selected_user = 69431
user = data[data.number == selected_user].reset_index(drop=True)
user.head()
```

```python
user = user.drop_duplicates('date').reset_index(drop=True)
```

```python
user['booking_class'] = 1
```

```python
date_range = pd.date_range(user.date.min(), user.date.max())
date_range
```

```python
user_missing_data = pd.DataFrame(date_range, columns=['date'])
user_missing_data.head()
```

```python
user_missing_data['booking_class'] = 0
user_missing_data['number'] = selected_user
user_missing_data.head()
```

```python
user_missing_data['is_weekend'] = user_missing_data.date.dt.dayofweek.isin([5,6])
user_missing_data['is_weekday'] = ~user_missing_data['is_weekend']
user_missing_data['is_start_of_week'] = user_missing_data.date.dt.dayofweek.isin([0,1])
user_missing_data['is_end_of_week'] = user_missing_data.date.dt.dayofweek.isin([3,4])
user_missing_data['is_wednesday'] = user_missing_data.date.dt.dayofweek.isin([2])
```

```python
user_missing_data.head()
```

```python
holidays = pd.read_csv('/home/logan/github/rapido-project-data/holiday.csv')
holidays.Date = pd.to_datetime(holidays.Date)
user_missing_data['is_holiday'] = user_missing_data.date.isin(holidays.Date)
```

```python
user_missing_data.head()
```

```python
user.head()
```

```python
user_missing_data = user_missing_data[~user_missing_data.date.isin(user.date)]
user_missing_data.head()
```

```python
user.dtypes
```

```python
user_missing_data.dtypes
```

```python
user.date = pd.to_datetime(user.date)
```

```python
complete_data = pd.merge(user, user_missing_data, how='outer')
complete_data.head()
```

```python
complete_data = complete_data.sort_values('date')
complete_data.head()
```

```python
complete_data.drop(['number','date'], inplace=True, axis=1)
complete_data.head()
```
